```python
# Simulação de Lógica de Backend Bancário
nome = input("Digite seu nome: ")
saldo = float(input("Quanto você tem na conta? R$ "))
saque = float(input("Quanto deseja sacar? R$ "))

if saque <= saldo:
    novo_saldo = saldo - saque
    print(f"Olá {nome}! Saque autorizado. Novo saldo: R$ {novo_saldo:.2f}")
else:
    print(f"Olá {nome}. Saldo insuficiente!")
```
